# Relatório de Análise Exploratória de Dados (EDA)

## Variáveis Numéricas

Esta seção **explora os padrões identificados em variáveis numéricas** e como essas variáveis afetam a probabilidade de compra de títulos. Os insights mostram o impacto de características numéricas, como faixa de renda e idade, na predisposição para investimentos, além de recomendar ajustes estratégicos nas abordagens de vendas e educação financeira para otimizar o atendimento. A análise inclui a recomendação de limites de contato com clientes, identificação de perfis com maior potencial e desenvolvimento de métricas de acompanhamento, visando melhorar a eficiência do processo de vendas e aumentar a taxa de conversão.

### **Faixa Etária**

**Predominância de Clientes na Faixa Etária de 25 a 40 Anos**: A maior parte dos clientes concentra-se entre 25 e 40 anos. Recomenda-se alinhar as estratégias de marketing e comunicação para atender melhor os interesses dessa faixa etária. Ações como questionários e entrevistas para identificar preferências financeiras e canais de comunicação podem ajudar a direcionar campanhas específicas e reforçar o relacionamento com esse público.

**Baixa Adesão à Compra de Títulos entre Clientes Potenciais**: Apesar do potencial de compra, muitos clientes de 25 a 40 anos não estão adquirindo títulos. Sugere-se investigar as barreiras para a compra por meio de grupos focais e entrevistas. Ações de conscientização e programas de educação financeira, como webinars e materiais explicativos, podem reduzir incertezas e incentivar a adesão a esses produtos.

**Necessidade de Estratégias Específicas para Clientes Acima de 40 Anos**: Clientes com mais de 40 anos apresentam necessidades e interesses financeiros distintos. Recomenda-se uma pesquisa detalhada para identificar produtos e serviços que atendam melhor a esse perfil, como opções de planejamento de aposentadoria e investimentos de baixo risco. Produtos personalizados e campanhas focadas em segurança e estabilidade podem gerar maior engajamento.

### **Situação Hipoteca**

**Predominância de Clientes Sem Hipoteca que Compram Títulos**: Observou-se que clientes sem hipoteca têm maior propensão para a compra de títulos. Recomenda-se investigar como a ausência de hipoteca influencia essa disposição para investir em produtos financeiros. Ações como a análise do perfil desses clientes e pesquisas sobre suas motivações, como segurança e diversificação financeira, podem ajudar a desenvolver estratégias focadas.

**Barreiras à Compra de Títulos Entre Clientes Sem Hipoteca**: Uma grande parcela de clientes sem hipoteca não está comprando títulos, o que indica a necessidade de abordagens específicas de conversão. Sugere-se uma análise aprofundada das razões por trás dessa resistência, através de entrevistas, questionários e análise de feedback. Esse processo ajudará a identificar e mitigar barreiras, como falta de conhecimento sobre investimentos.

**Clientes com Hipoteca que Compram Títulos**: Clientes com hipoteca também demonstram interesse na compra de títulos, indicando um perfil de investimento que busca equilibrar dívida e oportunidades financeiras. Recomenda-se investigar suas motivações e necessidades específicas para desenvolver produtos que ofereçam diversificação e suporte financeiro. Estratégias de pacotes de produtos e consultoria podem atender melhor a esse segmento e estimular novos investimentos.

### **Situação Devedor**

**Predominância de Clientes Não Devedores**: A base de clientes é majoritariamente composta por não devedores, indicando um perfil financeiro saudável. Recomenda-se explorar essa característica para promover produtos financeiros adequados a esse perfil, focando em oportunidades de investimento seguras. Ações como a realização de pesquisas e análises de comportamento podem ajudar a entender melhor as preferências desses clientes.

**Baixa Proporção de Clientes Devedores**: A proporção reduzida de clientes devedores sugere a oportunidade de implementar programas de educação financeira voltados a todos os clientes. É recomendada a criação de materiais que abordem a importância de uma gestão financeira equilibrada, incluindo os riscos do endividamento e as vantagens de manter um perfil saudável.

**Oportunidade de Atração de Novos Clientes**: Existe potencial para atrair clientes com perfil financeiro saudável, semelhante à atual base de clientes. Recomenda-se focar em estratégias de aquisição que destaquem a reputação do banco e a acessibilidade de seus produtos, atraindo indivíduos com um perfil de baixo risco.

### **Quantidade de Ligações**


**Diminuição de Conversões a Partir da Sexta Ligação**: Foi observada uma queda significativa nas taxas de conversão após a sexta ligação. Recomenda-se, portanto, a implementação de um limite de seis tentativas de contato por cliente, otimizando a alocação de recursos da equipe de vendas e aumentando o foco nos leads mais responsivos.

**Análise de Desempenho**: Conduzir uma análise aprofundada para mapear a relação entre o número de ligações e a taxa de conversão. Esse mapeamento deve identificar padrões de conversão nas interações iniciais, permitindo definir um ponto de corte eficiente para o número de ligações.

**Diretrizes para a Equipe de Vendas**: Baseando-se nos insights obtidos, estabelecer diretrizes de vendas que orientem a equipe sobre a alocação ideal de tempo e recursos. A equipe deve receber treinamento para maximizar as conversões nas primeiras interações, com foco em leads de maior potencial, e usar técnicas que aumentem as chances de fechamento nas primeiras tentativas.

## Variáveis Categóricas

Este tópico **examina os padrões identificados entre variáveis categóricas** e a relação desses fatores com a probabilidade de compra de títulos pelo cliente. São abordados insights sobre a influência do nível educacional na propensão a investir, o perfil de clientes com alta taxa de conversão, como estudantes e aposentados, e as oportunidades de personalizar campanhas para diferentes grupos. A análise propõe estratégias de educação financeira, campanhas direcionadas e programas de fidelização, com o objetivo de reforçar a familiaridade com produtos financeiros e promover o aumento nas conversões.

**Impacto do Nível Educacional na Propensão a Investir**: Observou-se que clientes com menor escolaridade apresentam menor propensão à compra de títulos. Recomenda-se a criação de programas educativos adaptados, que incentivem a compreensão e o acesso aos produtos financeiros do banco.

**Educação Financeira para Clientes de Baixa Escolaridade**: Desenvolver workshops, cursos online e materiais que expliquem conceitos básicos de investimentos, com linguagem acessível e exemplos práticos. Materiais explicativos em vídeo, folhetos e e-books devem ser incluídos para alcançar uma ampla compreensão dos produtos e suas vantagens.

**Alta Conversão Entre Estudantes**: Estudantes apresentam a maior taxa de conversão para produtos financeiros. Recomenda-se a criação de campanhas focadas em planejamento financeiro de longo prazo, utilizando temas relevantes para esse grupo.

**Conversão Abaixo da Média em Grupos de Baixa Renda**: Clientes com menor renda apresentam uma taxa de conversão reduzida. A proposta é adaptar produtos financeiros de baixo risco e alta liquidez para atrair esse público. Podem ser feitas pesquisas para entender melhor as necessidades financeiras desse grupo e ajustar os produtos para atender a essas especificidades. A criação de produtos com benefícios claros, como segurança e liquidez, pode atrair clientes de renda mais baixa.

**Aposentados, um Perfil com Alta Conversão**: Clientes aposentados têm maior propensão a adquirir produtos financeiros. Recomenda-se a oferta de produtos de investimento com foco em segurança e preservação do patrimônio. É importante utilizar newsletters, eventos presenciais e comunicações adaptadas aos canais mais comuns para aposentados, garantindo que a mensagem atinja o público-alvo de forma eficaz.

**Administração, um Segmento Importante na Base de Clientes**: O grupo de administradores representa uma parcela significativa de clientes. Recomenda-se a análise aprofundada das necessidades financeiras desse público para desenvolver produtos que atendam a suas expectativas.

**Oportunidade de Educação Financeira Segmentada**: Há uma demanda crescente por educação financeira personalizada entre clientes com diferentes perfis. Sugere-se a implementação de um programa abrangente que facilite o entendimento dos investimentos de longo prazo.
